Considerando um dataset onde eu tenho as variáveis explicativas do fenômeno/problema e a variável alvo, primeiro temos que analisar qual será o limiar de decisão para :

1. Tratar os valores ausentes de uma determinada variável explicativa ou 
2. descartar essa variável pois atingiu o limite, ou seja, possui muitos valores ausente, não é válido para o estudo manter essa variável, visto que, para manter ela teria que gerar um impultação de dados muito alta, o que pode ser um problema para a aplicação do modelo em produção, por que, depende claro da variável, não sabemos se o valor atribuido é de fato aquele valor para um determinado caso.

Logo, esse limiar poderia ser 60% ou 80%

considerando uma variável em que é possível realizar o tratamento, está por exemplo com 32% de valores ausentes, temos algumas possibilidades para garantir que aquela linha nao seja perdida:

1. realizar a substituição pela média e essa pode ser uma média geral dependendo se a variável apresenta ou não muito outliers, visto que, a presença de outiliers tem alto impacto no valor da média. Ou pode ser uma média em conjunto, considerando outra variável para forma um grupo de médias e substituir, por exemplo em um conjunto de dados em que fala de edifícios e tenho outra variável que classifica eles pelo tamanho assim eu posso fazer médias em relação a grupos de tamanhos e depois realziar a substituição

2. Podemos utilizar a mediana também, essa é menos sentível com relação a outiliers
3. Podemos utilzar para casos onde temos valores faltantes em variável categóricas, a imputação pela moda
4. O que émuito utilizado também é o uso de modelo para encontrar o melhor valor para substituir aquele valor ausente, por exemplo, utilizar um random florest para essa tarefa.

5. Também temos que conhecer o conjunto de dados, e verificar com a área de negócio se o valor ausente represente ou tem alguma representatividade no estudo. É esperado aquele valor ausente? Se sim, realizar por exemplo a substituição com um valor no qual o modelo vai conseguir interpretar (-1)



Enfim, realizar o tratamento e a preparação dos dados é uma etapa super importante, arrisco dizer que é uma das etapas mais prolongadas em um projeto, visto que, o desempenho do modelo vai depender muito da forma como o dataset foi tratado, como foi aproveitado as variáveis, as decisões tomadas.

O que é uma matriz de confusão e como ela é usada para avaliar o
desempenho de um modelo preditivo?
</p>

A matriz de confusão ela é uma ferramenta de avaliação utilizada para modelos preditivos de classificação, binária ou multiclasses, que está baseado da seguinte forma: 

Nessa representação da matriz de confusão, temos uma estrutura de tabela/matriz, onde inicialmente temos no eixo vertical destacado com o nome real, o valor real da variável alvo, logo aquele vai ser o eixo para os valores reais os valores que já vieram com o rótulados com o estudo, no eixo horizontal temos os valores que foram preditos pelo modelo de classificação

<div style="border-left: 6px solid #FF9800; background-color: #f9f9f9; padding: 15px; border-radius: 10px; box-shadow: 0px 2px 4px rgba(0,0,0,0.1);">
  <h3 style="margin: 0; color: #FF5722;">⚠️ Observações:</h3>

  <p style="margin-top: 5px; color: rgb(0, 0, 0);">
   - Essa ordem de eixo pode variar com relação à decisão do programador que estar montado a matriz
  </p>
</div>


No meio da matriz de confusão temos as interrações entre os valores reais e os valores preditos das classes, ou seja, um valor no qual é rótulado como Sim no eixo real e o valor obteve com a predição o mesmo rótulo Sim, esse significa que o modelo acertou, ele é um verdadeito (acerto) positivo (pois representa o acontecimento do fato, colocado como 1)

Resumo:

1. Verdadeiro Positivo (VP)
O modelo previu positivo e estava certo.

Exemplo: o modelo disse que era fraude e era realmente fraude.

2. Falso Positivo (FP)
O modelo previu positivo, mas estava errado.

Exemplo: o modelo disse que era fraude, mas não era fraude.

Também chamado de erro tipo I.

3. Falso Negativo (FN)
O modelo previu negativo, mas estava errado.

Exemplo: o modelo disse que não era fraude, mas era fraude.

Também chamado de erro tipo II.

4. Verdadeiro Negativo (VN)
O modelo previu negativo e estava certo.

Exemplo: o modelo disse que não era fraude e realmente não era fraude.

<div style="border-left: 6px solid #FF9800; background-color: #f9f9f9; padding: 15px; border-radius: 10px; box-shadow: 0px 2px 4px rgba(0,0,0,0.1);">
  <h3 style="margin: 0; color: #FF5722;">⚠️ Observações:</h3>

  <p style="margin-top: 5px; color: rgb(0, 0, 0);">
   - Podemos identificar que a diagonal principal da matriz de confusão representa os valores que o modelo conseguiu acerta
   - Essa característica de construção da matriz vai se expandir para uma classificação de multiplas classes,
  </p>
</div>


Desse modo, já temos então uma ferramente que visualmente conseguimos retirar uma inferência, por mais que visual, a principio, se o modelo está ou não acertando, se ele está errando muito, se ele está errando muito uma determinada classe.

E a matriz de confusão é uma ferramente poderosa para preparar o modelo da melhor forma para atender o problema do negócio, pois através dos valores que temos de (VP, FP, FN, VN), podemos obter métricas nos quais podemos decidir qual o melhor valor para cada métrica em relação ao estudo

### **Acurácia**
Mede a proporção de acertos (tanto positivos quanto negativos) sobre todas as previsões.
$$
\text{Acurácia} = \frac{VP + VN}{VP + FP + FN + VN}
$$

Interpretação: Quantas vezes o modelo acertou no total.



### **Precisão**
Mede a proporção de acertos entre todas as vezes que o modelo previu positivo.
$$
\text{Precisão} = \frac{VP}{VP + FP}
$$
Interpretação: Quando o modelo disse que era positivo, quantas vezes ele estava certo?

### **Revocação (Recall ou Sensibilidade)**

Mede a proporção de positivos corretamente identificados.
$$
\text{Revocação} = \frac{VP}{VP + FN}
$$
Interpretação: Entre todos os casos realmente positivos, quantos o modelo conseguiu encontrar?

### **Especificidade (Specificity)**
Mede a proporção de negativos corretamente identificados.
$$
\text{Especificidade} = \frac{VN}{VN + FP}
$$
Interpretação: Entre todos os casos realmente negativos, quantos o modelo identificou corretamente?



### **F1-Score**

É a média harmônica entre Precisão e Revocação.

Equilibra a importância entre acertar os positivos e encontrá-los.
$$
F1 = 2 \times \frac{\text{Precisão} \times \text{Revocação}}{\text{Precisão} + \text{Revocação}}
$$
Interpretação: Uma métrica única que combina Precisão e Revocação.

### **Taxa de Falsos Positivos (FPR)**

Mede a proporção de negativos incorretamente classificados como positivos.
$$
\text{Taxa de Falsos Positivos} = \frac{FP}{FP + VN}
$$

Interpretação: Entre os casos que eram negativos, quantos o modelo classificou errado como positivos?
